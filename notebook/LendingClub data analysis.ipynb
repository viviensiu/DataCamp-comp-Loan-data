# Background
<i> Excerpts from The LendingClub's Wikipedia </i>

The LendingClub is a fintech that provides range of financial products and services through a technology-driven platform in the United States.

It enables borrowers to create unsecured personal loans between 1,000 USD and 40,000 USD. On the basis of the borrower’s <strong>credit score, credit history, desired loan amount and the borrower’s debt-to-income ratio </strong>, LendingClub determined whether the borrower was creditworthy and assigned to its approved loans a <strong>credit grade</strong> that determined the payable interest rate and fees. The standard loan period is three years. 

Investors are able to search and browse the loan listings on LendingClub website and select loans that they want to invest in based on the information supplied about the <strong>borrower, amount of loan, loan grade, and loan purpose</strong>. Institutional investors make money from the interest on these loans. Rates varied from 6.03% to 26.06%, depending on the <strong>credit grade</strong> assigned to the loan request.

# Analysing variables for investment risks
Let's take a look next at the following variables that may pose risks to our investments:

- days_with_cr_line: number of days borrower has had a credit line.
- revol_bal: borrower's revolving balance (amt unpaid at the end of credit card billing cycle).
- revol_util: borrower's revolving line utilization rate (amt of credit line used relative to total credit available).
- inq_last_6_mths: borrower's number of inquiries by creditors in last 6 months.
- delinq_2yrs: number of times that borrower has been 30+ days past due on a payment in past 2 years.
- pub_rec: borrower's number of derogatory public record.

We will plot them against the Credit Policy variable as it indicates if the credit underwriting gets approved or otherwise.

Read more about Credit Underwriting here: <a href='https://www.rocketmortgage.com/learn/what-is-underwriting'> What is Underwriting?</a>

## Credit Policy vs Days with Cr Line

We will look at Revolving Balance next.

## Credit Policy vs Revolving Balance

```python
sns.displot(df, x='revol_bal', hue='credit_policy', kind='kde')
plt.title('Credit Policy vs Revolving Balance distribution')
plt.show()
```

It's hard to make sense of the distribution above as the distribution is highly imbalanced with outliers. Therefore we need to filter the revolving balance data first to plot only the data that falls within the interquartile range.

```python
Q3 = np.quantile(df['revol_bal'], 0.75)
Q1 = np.quantile(df['revol_bal'], 0.25)
IQR = Q3 - Q1

lower_range = Q1 - 1.5 * IQR
upper_range = Q3 + 1.5 * IQR
outlier_free_list = [x for x in df['revol_bal'] if ((x > lower_range) & (x < upper_range))]
filtered_data = df.loc[df['revol_bal'].isin(outlier_free_list)]
```

```python
sns.displot(filtered_data, x='revol_bal', hue='credit_policy', kind='kde')
plt.subplots_adjust(top=0.9)
plt.suptitle('Credit Policy vs Revolving Balance IQR distribution')
plt.show()
```

Well, seems like having a Revolving Balance close to zero does not imply your loan underwriting will get approved either.

## Credit Policy vs Revol. Line Util.

## Factors on maximizing profits
As our data did not include the total loan amount, we will look at interest rates and installments mainly for the potential returns.

First, we look at <strong>FICO scores</strong> because it have a strong negative correlation with interest rates.
The FICO scores have a direct impact on the interest rates. Higher FICO score = lower interest rates. 

Next, <strong>loan purposes </strong>.
We will prioritize loans for Debt Consolidation, Credit Card, all others, home improvements and small business. These loans typically generate an interest rates of 6% - 14% (for small business loans it can go up to 16%), and there are potential loans within this group that could perform well above the mentioned range.

<strong>Annual Income</strong>
It seems that people with more income can afford to take out larger loans, indicating potentially better returns. These people generally fall in the bracket of log annual income of 10 and higher.


## Factors on minimizing risks

Ironically, we also need to look at <strong>FICO scores</strong>.
A borrower's FICO scores is typically influenced by their past credit history. If they're diligent on payments, FICO scores are high, lenders don't feel the need to charge higher interests to mitigate risks.

<strong>Number of 30+ days past due on payments</strong>
Majority of approved loans do not delay payments for more than 30+ days. However we understand that sometimes accidents do occur, hence we will set a limit of up to 1 for this variable.

<strong>Derogatory public records</strong>
Public records = bankruptcies. Anything above zero is a nope.

<strong>Number of inquiries in past 6 months</strong>
We will limit the number to 3 and below since loan approvals start to drop significantly past this number.

<strong>Revolving Line Utilization</strong>
We will limit this to 100% as overdrawn credit is a bad indication on how well the borrower manages their expenses.

<strong>Annual Income</strong>
Not only the higher income bracket generates potentially better loans, but they run a lower risk of defaulting their loans. For this reason, we will look at borrowers with log annual income of 10 and higher.

# Outcome
Let's apply our analysis results onto the dataset and see what the outcome is! We will filter the dataset based on the factors mentioned and see how our investment turns out!

```python
filter_past_due = (df['delinq_2yrs'] <= 1)
filter_pub_rec = (df['pub_rec'] == 0)
filter_inq_last_6mths = (df['inq_last_6mths'] <= 3)
filter_revol_util = (df['revol_util'] <= 100)
filter_income = (df['log_annual_inc'] >= 10)

df_filtered = df[filter_past_due &
                 filter_pub_rec &
                 filter_inq_last_6mths &
                 filter_revol_util &
                 filter_income]
                 
df_filtered.describe()
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax = sns.boxplot(data=df_filtered, x='purpose', y='int_rate')
plt.title('Interest Rates interquartile range across Loan Purposes')
plt.xticks(rotation=90)
plt.show()

total_count = df_filtered.shape[0]
fully_paid = df_filtered[df_filtered.not_fully_paid == 0]['not_fully_paid'].count()
not_fully_paid = df_filtered[df_filtered.not_fully_paid == 1]['not_fully_paid'].count()

pct_fully_paid = (fully_paid/total_count) * 100
pct_not_fully_paid = (not_fully_paid/total_count) * 100

print('Out of {} borrowers,'.format(total_count))
print('{:.2f}% paid their loans in full.'.format(pct_fully_paid))
print('{:.2f}% did not fully pay their loans.'.format(pct_not_fully_paid))
```

The new data set now has 7180 records (74.9%) as compared to original number of 9578 records.

Compared to the original data set, our filtered data have similar interest rates where the interquartile range lies between 9% to 13%. Again, small business loans have a better IQR range of rates between 11% to a little under 16%. 

We also noticed fewer outliers in this data set once we factor in the risks. Goes to show that greater interest rates also implies higher risks of losing your investments!

One more thing, despite having a reduced pool size of loans to invest, we now face a much higher percentage of loans paid in full (86.79%). This is a positive increase from the initial 83.99% in the original data set!

# Conclusion and some thoughts on this data set
<strong>Ultimately,</strong> investment is all about how much you want your returns to be, and how much risks you're willing to take.

Taking a moderate, calculated risks' approach did not reduce our returns compared to the general statistics and we still have some attractive potential high returns choices to go for. On top of this, we have a greater chance of getting good borrowers who pay back in full.

On the other hand, we have seen that going for high returns alone carries an increased risk of losing our money.

Looking at the various LendingClub investment resources, this data set could benefit with data enrichment as it's missing some important variables, like <strong>loan grade, loan amount, and loan terms,</strong> which will really help us in performing further feature engineering for gauging on the actual returns.
